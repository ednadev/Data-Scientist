```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("../data/02. sales-funnel.xlsx")
```

```python
df.head(3)
```

```python
# 1. 피벗 기능을 이용하여 아래 자료를 표현하세요.
df.pivot_table(index='Name')
```

```python
# 2. 피벗 기능을 이용하여 아래 자료를 표현하세요.
df.pivot_table(index=['Name', 'Rep', 'Manager'])
```

```python
# 3. 피벗 기능을 이용하여 아래 자료를 표현하세요. (기본적인 값은 price가 평균값으로 계산됩니다)
df.pivot_table('Price', index=['Manager', 'Rep'])
```

```python
# 4. 피벗 기능을 이용하여 아래 자료를 표현하세요. 
df.pivot_table('Price', index=['Manager', 'Rep'], aggfunc='sum')
```

```python
# 5. 피벗 기능을 이용하여 아래 자료를 표현하세요.
df.pivot_table(['Price', 'Quantity'], index=['Manager', 'Rep', 'Product'], aggfunc=['sum', 'mean'], margins=True).style.set_properties(**{'text-align': 'right'})
```
