### Plot 직선그리기

```python
# 1. plot() : 직선, 혹은 꺽은선 형태의 그래프를 그릴 수 있다...선형 | 비선형
# 간단한 데이타를 하나 만들어서..(x,y) 직선 혹은 꺽은선을 그려보자
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # 첫번째 리스트값 X, 두번째 리스트값 Y
plt.show()

# 2. 그렇다면 직선은?
plt.plot([1, 2, 3, 4], [1, 2, 3, 4]) 
plt.show()

# 3.리스트에 입력한 값은 Y축의 값으로 입력되며, X축 값은 자동으로 0에서 1씩 증가
plt.plot([10, 20, 30, 40])
plt.show()
```

### 겹쳐진 그래프 그리기

```python
import numpy as np
```

```python
# 0부터 5미만까지 0.2간격으로 출력
t = np.arange(0, 5, 0.2)
t

plt.plot(t, t, 'g--')
plt.show()
```

```python
plt.plot(t, t, 'r--', t, t**2, 'gs') # s는 스퀘어..
plt.show()
```

```python
plt.plot(t, t, 'r--', t, t**2, 'gs', t, t**3, 'b^') # s는 스퀘어..
plt.show()
```

```python
plt.plot(t, t, 'r--', label='Tips') 
plt.plot(t, t**2, 'gs', label='Population') 
plt.plot(t, t**3, 'b^', label='Product Layer') 

plt.legend() # 이부분을 해야 label표식이 나타난다.
plt.title('Wrapped Graphs')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid()
plt.show()
```

### 파도 모양의 그래프

```python
t = np.arange(0, 12, 0.01)
'''
기본으로 존재하는 figure가 아닌 새로운 figure를 생성하기 위해서는
plt.figure() 함수를 사용한다.
'''
plt.figure(figsize=(8, 5))

plt.plot(t, np.sin(t), 'r', lw=10, label='sin') # line width, 기본값이 3
plt.plot(t, np.cos(t), 'g', label='cos')

plt.legend()
plt.xlabel('Time')
plt.ylabel('Amplitute')
plt.title('SinCos Graph')
plt.show()
```

### Pandas 데이타를 이용한 Visualization

```python
np.random.seed(100)
#1. 딕셔너리 구조로 데이타프레임 생성
data = {'a':np.arange(50), 
        'c':np.random.randint(0, 50, 50),
        'd':np.random.randn(50)}
data
```

```python
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d'])*100

data
```

```python
import pandas as pd
from pandas import DataFrame

DataFrame(data)
```

### 산점도(Scatter) 그래프 그리기

```python
'''
산점도 그래프는 scatter() 함수를 이용해서 그린다.
산점도는 x,y축에 해당하는 데이타들의 상관관계를 표시할때 데이타들이
어떻게 흩어져 있는가를 알기위해서 사용된다.

2개의 축을 기준으로 데이타가 얼마나 펴져있는지(분포도)를 알수 있기 때문에
산포도 혹은 산점도로 불린다. 
'''
```

```python
# 많은 양의 데이터를 사용할 때는 s와 c는 사용하지 X
plt.scatter([1, 2, 3, 4], [10, 20, 30, 40], s=[100,200,300,400], c=['yellow', 'green', 'magenta', 'red'])
plt.show()

plt.scatter([1, 2, 3, 4], [10, 20, 30, 40], s=[100,200,300,400], c=range(4))
plt.show()

plt.scatter([1, 2, 3, 4], [10, 20, 30, 40], s=[100,200,300,400], c=range(4), cmap='jet')
plt.colorbar()
plt.show()
```

```python
plt.scatter('a', 'b', data = data, c = 'c')
plt.show()
```

```python
import seaborn as sns

cuthalf = (data['a']>25) & (data['b']>30)
data['color'] = np.where(cuthalf==True, 'red', 'blue')

sns.regplot(x=data['a'], 
            y=data['b'], 
            fit_reg=False,
            scatter_kws={'facecolors':data['color']})
plt.title('Scatter Plot with Colors', fontsize=20)
plt.show()
```

### Subplot으로 plot(), scatter(), bar() 동시 이용하기

```python
names = ['group_A', 'group_B', 'group_C'] #x축으로 사용
values = [1, 10, 100] #y축으로 사용
```

```python
plt.figure(1, figsize=(9,3)) # figsize=(9,3)영역이 첫번째 라는 의미로 1

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)

plt.show()
```
